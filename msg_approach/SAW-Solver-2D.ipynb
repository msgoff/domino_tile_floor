```python
def traversal_(r,c):
    dct={}
    bndry = df.shape
    #print bndry
    if (r >=0 and r< bndry[0]):
        dct[r]=[x for x in [c-1,c+1] if (x >=0 and x< bndry[1])]
        
    if (r-1 >=0 and r-1< bndry[0]): 
        dct[r-1]=[x for x in [c-1,c,c+1] if (x >=0 and x< bndry[1])]
    
    if (r+1 >=0 and r+1< bndry[0]):
        dct[r+1]=[x for x in [c-1,c,c+1] if (x >=0 and x< bndry[1])]
    
        
    return dct
```
